# Data Story: Xu hướng tiền tips và tổng hóa đơn theo ngày trong tuần


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("tips.csv")

# Đảm bảo cột 'day' có thứ tự đúng
order = ["Sun", "Mon", "Tue", "Wed", "Thu", "Fri", "Sat"]
df["day"] = pd.Categorical(df["day"], categories=order, ordered=True)

# Tính tổng tips và tổng hóa đơn theo ngày
grouped = df.groupby("day", observed=False)[["total_bill", "tip"]].sum()

# Vẽ biểu đồ 2 series
ax = grouped.plot(kind="line", marker="o")
ax.set_xlabel("Ngày trong tuần")
ax.set_ylabel("Tổng giá trị ($)")
ax.set_title("Xu hướng tổng hóa đơn và tiền tips theo ngày trong tuần")
ax.legend(["Tổng hóa đơn", "Tổng tips"])
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

Biểu đồ trên minh họa tổng số tiền tips và tổng hóa đơn mà nhà hàng nhận được trong từng ngày của tuần. Dưới đây là một số điểm nổi bật từ dữ liệu:

- **Ngày cuối tuần (Chủ nhật)** thường có tổng hóa đơn và tiền tips cao nhất, cho thấy lượng khách đông và chi tiêu nhiều hơn vào dịp này.
- **Các ngày giữa tuần** như Thứ Hai đến Thứ Năm có tổng hóa đơn và tips thấp hơn đáng kể, phản ánh xu hướng nhà hàng vắng khách hơn so với cuối tuần.
- **Thứ Sáu và Thứ Bảy** cũng ghi nhận mức tăng nhẹ về tổng hóa đơn và tips, cho thấy khách hàng bắt đầu đi ăn ngoài nhiều hơn khi gần đến cuối tuần.
- **Tương quan giữa tổng hóa đơn và tips**: Biểu đồ cho thấy khi tổng hóa đơn tăng thì tổng tips cũng tăng theo, phản ánh mối liên hệ tích cực giữa chi tiêu của khách và tiền thưởng cho phục vụ.
