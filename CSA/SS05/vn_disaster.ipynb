# Phân tích thảm họa công nghệ tại Việt Nam dựa trên dữ liệu thực tế

Dưới đây là phân tích các thảm họa công nghệ (tai nạn công nghiệp, tai nạn giao thông, hỏa hoạn...) tại Việt Nam dựa trên file `vn_disaster.csv`.


```python
import pandas as pd

df = pd.read_csv("vn_disaster.csv")
# Chỉ hiển thị các cột liên quan đến thảm họa công nghệ
cols = [
    "Disaster Group",
    "Disaster Type",
    "Disaster Subtype",
    "Event Name",
    "Start Year",
    "Location",
    "Total Deaths",
    "Total Damage ('000 US$)",
]
print(f"Dữ liệu:\n {df[cols].head()}\n")

# Lọc các thảm họa công nghệ (Technological)
tech_df = df[df["Disaster Group"] == "Technological"]

# Thống kê số lượng từng loại thảm họa công nghệ
tech_types = tech_df["Disaster Type"].value_counts()
print("Số lượng từng loại thảm họa công nghệ:")
print(tech_types)

# Thống kê tổng số người chết do thảm họa công nghệ
total_deaths = tech_df["Total Deaths"].sum()
print("\nTổng số người chết do thảm họa công nghệ:", total_deaths)

# Thống kê tổng thiệt hại (nếu có dữ liệu)
if "Total Damage ('000 US$)" in tech_df.columns:
    total_damage = tech_df["Total Damage ('000 US$)"].sum()
    print("\nTổng thiệt hại (nghìn USD):", total_damage)
```

## Nhận xét và đề xuất giảm thiểu thảm họa công nghệ

- **Các loại thảm họa công nghệ phổ biến nhất** thường là tai nạn giao thông (đường bộ, đường thủy, đường hàng không), tai nạn công nghiệp, hỏa hoạn và ngộ độc.
- **Thiệt hại về người và tài sản** do các thảm họa này gây ra là rất lớn, đặc biệt ở các khu vực đông dân cư và các ngành công nghiệp trọng điểm.

### Đề xuất giảm thiểu:

1. **Nâng cao tiêu chuẩn an toàn lao động và công nghiệp**: Tăng cường kiểm tra, giám sát quy trình sản xuất, bảo trì thiết bị.
2. **Đào tạo và nâng cao nhận thức** cho người lao động, lái xe, cán bộ quản lý về phòng tránh tai nạn.
3. **Cập nhật và thực thi nghiêm ngặt các quy định pháp luật** về an toàn giao thông, phòng cháy chữa cháy, an toàn hóa chất.
4. **Ứng dụng công nghệ giám sát, cảnh báo sớm** trong các lĩnh vực có nguy cơ cao.
5. **Phân tích dữ liệu thường xuyên** để nhận diện các điểm nóng, xu hướng mới và kịp thời điều chỉnh chính sách.

Việc sử dụng dữ liệu thực tế giúp các cơ quan chức năng đưa ra quyết định chính xác, chủ động phòng ngừa và giảm thiểu hậu quả của thảm họa công nghệ.
