# Phân tích về giá nhà của một công ty định giá


---


### Dữ liệu


```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression

df = pd.read_csv('housing_data.csv')
df.head()
```

---


### Xây dựng mô hình


```python
model = LinearRegression()
x = df[['Size', 'Age']]
y = df['Price']
model.fit(x, y)
```

---


#### Chỉ số R<sup>2</sup>


```python
model.score(x, y)
```

Điều này cho thấy mô hình có độ hiệu quả rất cao, ở mức khoảng 98,54%.


---


### So sánh mối tương quan của Diện tích và Tuổi với Giá nhà


```python
plt.scatter(df['Size'], df['Price'])
plt.xlabel('Size (m²)')
plt.ylabel('Price (000$)')
plt.title('Size vs Price')
plt.show()
```

```python
plt.scatter(df['Age'], df['Price'])
plt.xlabel('Age (years)')
plt.ylabel('Price (000$)')
plt.title('Age vs Price')
plt.show()
```

Ta thấy:

- Biểu đồ so sánh yếu tố Diện tích và Giá nhà có mối tương quan rõ rệt.
- Biểu đồ so sánh yếu tố Tuổi và Giá nhà gần như không có mối tương quan.
