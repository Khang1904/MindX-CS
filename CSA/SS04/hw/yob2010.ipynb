```python
# Import th∆∞ vi·ªán v√† d·ªØ li·ªáu
import pandas as pd

data = pd.read_csv("yob2010.csv")
```

## üü£ C√¢u h·ªèi 1

**C√≥ bao nhi√™u em b√© trai v√† em b√© g√°i ƒë∆∞·ª£c sinh v√†o nƒÉm 2010?**


```python
# T·ªïng s·ªë b√© trai v√† b√© g√°i sinh nƒÉm 2010
gender_count = data.groupby("Gender")["Count"].sum()
print(gender_count)
```

## üü£ C√¢u h·ªèi 2

**Top 10 t√™n ph·ªï bi·∫øn nh·∫•t ƒë·∫∑t cho b√© g√°i v√† b√© trai l√† g√¨?**


```python
# Top 10 t√™n ph·ªï bi·∫øn nh·∫•t cho b√© g√°i
f_top10 = (
    data[data["Gender"] == "F"]
    .groupby("Name")["Count"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
print(f_top10)

# Top 10 t√™n ph·ªï bi·∫øn nh·∫•t cho b√© trai
m_top10 = (
    data[data["Gender"] == "M"]
    .groupby("Name")["Count"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
print(m_top10)
```
