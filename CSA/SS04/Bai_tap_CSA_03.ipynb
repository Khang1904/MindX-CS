## **I. Chu·∫©n B·ªã D·ªØ Li·ªáu (Data Loading & Inspection)**


### **1. Gi·ªõi thi·ªáu**

ƒê·∫ßu ti√™n, ch√∫ng ta c·∫ßn n·∫°p d·ªØ li·ªáu t·ª´ file `hw3_data.csv` v√†o m·ªôt **DataFrame** (t∆∞·ªüng t∆∞·ª£ng l√† m·ªôt c√°i b·∫£ng t√≠nh Excel trong Python) ƒë·ªÉ c√≥ th·ªÉ thao t√°c d·ªÖ d√†ng.


### **2. Code: N·∫°p v√† Ki·ªÉm tra d·ªØ li·ªáu**

S·ª≠ d·ª•ng th∆∞ vi·ªán **pandas** (th∆∞·ªùng ƒë∆∞·ª£c vi·∫øt t·∫Øt l√† `pd`) ƒë·ªÉ l√†m vi·ªác n√†y!


```python
import pandas as pd

# 1. ƒê·ªçc d·ªØ li·ªáu t·ª´ file CSV
df = pd.read_csv("hw3_data.csv")

# 2. Xem 5 d√≤ng ƒë·∫ßu ti√™n ƒë·ªÉ hi·ªÉu c·∫•u tr√∫c b·∫£ng
print("--- 5 D√≤ng ƒë·∫ßu ti√™n c·ªßa DataFrame ---")
print(df.head().to_markdown(index=False, numalign="left", stralign="left"))

# 3. Xem th√¥ng tin c√°c c·ªôt (ki·ªÉu d·ªØ li·ªáu, s·ªë l∆∞·ª£ng d√≤ng)
print("\n--- Th√¥ng tin c√°c c·ªôt d·ªØ li·ªáu (df.info()) ---")
df.info()
```

## **II. Gi·∫£i Quy·∫øt C√°c C√¢u H·ªèi**


### **C√¢u h·ªèi 1: C√≥ bao nhi√™u ng∆∞·ªùi tr√™n 30 tu·ªïi trong d·ªØ li·ªáu?**


#### **üí° √ù t∆∞·ªüng:**

Ch√∫ng ta s·∫Ω s·ª≠ d·ª•ng k·ªπ thu·∫≠t **l·ªçc Boolean (Boolean Indexing)** c·ªßa Pandas. ƒêi·ªÅu ki·ªán l·ªçc l√†: **c·ªôt 'Age' ph·∫£i l·ªõn h∆°n** $30$ ($\text{df}[\text{'Age'}] > 30$). Sau ƒë√≥, ƒë·∫øm s·ªë l∆∞·ª£ng d√≤ng th·ªèa m√£n.


#### **üõ†Ô∏è Code:**


```python
# L·ªçc: T·∫°o m·ªôt DataFrame m·ªõi ch·ªâ ch·ª©a nh·ªØng ng∆∞·ªùi c√≥ Age > 30
people_over_30 = df[df["Age"] > 30]

# ƒê·∫øm s·ªë l∆∞·ª£ng d√≤ng trong DataFrame ƒë√£ l·ªçc
count_over_30 = len(people_over_30)

print(f"C√¢u tr·∫£ l·ªùi 1: C√≥ {count_over_30} ng∆∞·ªùi tr√™n 30 tu·ªïi trong d·ªØ li·ªáu.")
```

**K·∫øt qu·∫£ 1:** C√≥ **3** ng∆∞·ªùi tr√™n 30 tu·ªïi.


---


### **C√¢u h·ªèi 2: T·ªïng s·ªë ti·ªÅn l∆∞∆°ng c·ªßa t·∫•t c·∫£ c√°c ng∆∞·ªùi c√≥ trong d·ªØ li·ªáu l√† bao nhi√™u?**


#### **üí° √ù t∆∞·ªüng:**

ƒê·ªÉ t√≠nh t·ªïng, ch√∫ng ta ch·ªâ c·∫ßn ch·ªçn c·ªôt **'Salary'** v√† √°p d·ª•ng h√†m `.sum()` n·ªïi ti·∫øng c·ªßa Pandas.


#### **üõ†Ô∏è Code:**


```python
# T√≠nh t·ªïng gi√° tr·ªã c·ªßa to√†n b·ªô c·ªôt 'Salary'
total_salary = df["Salary"].sum()

print(
    f"C√¢u tr·∫£ l·ªùi 2: T·ªïng s·ªë ti·ªÅn l∆∞∆°ng c·ªßa t·∫•t c·∫£ m·ªçi ng∆∞·ªùi l√† {total_salary:,} (VNƒê/ƒê∆°n v·ªã ti·ªÅn t·ªá)."
)
```

**K·∫øt qu·∫£ 2:** T·ªïng s·ªë ti·ªÅn l∆∞∆°ng l√† **430,000** (VNƒê/ƒê∆°n v·ªã ti·ªÅn t·ªá).


---


### **C√¢u h·ªèi 3: C√≥ bao nhi√™u ng∆∞·ªùi c√≥ l∆∞∆°ng l·ªõn h∆°n 50,000 v√† tu·ªïi nh·ªè h∆°n 40?**


#### **üí° √ù t∆∞·ªüng:**

ƒê√¢y l√† c√¢u h·ªèi l·ªçc d·ªØ li·ªáu v·ªõi **hai ƒëi·ªÅu ki·ªán** c√πng l√∫c. Ch√∫ng ta c·∫ßn k·∫øt h·ª£p hai ƒëi·ªÅu ki·ªán b·∫±ng ph√©p to√°n **AND** (k√Ω hi·ªáu l√† `&` trong Pandas) v√† nh·ªõ bao m·ªói ƒëi·ªÅu ki·ªán trong d·∫•u ngo·∫∑c ƒë∆°n `()`.

1. ƒêi·ªÅu ki·ªán 1: $\text{df}[\text{'Salary'}] > 50000$
2. ƒêi·ªÅu ki·ªán 2: $\text{df}[\text{'Age'}] < 40$
3. K·∫øt h·ª£p: $(\text{df}[\text{'Salary'}] > 50000) \boldsymbol{\&} (\text{df}[\text{'Age'}] < 40)$


#### **üõ†Ô∏è Code:**


```python
# L·ªçc: K·∫øt h·ª£p hai ƒëi·ªÅu ki·ªán (Salary > 50000) V√Ä (Age < 40)
people_filtered = df[(df["Salary"] > 50000) & (df["Age"] < 40)]

# ƒê·∫øm s·ªë l∆∞·ª£ng d√≤ng th·ªèa m√£n
count_filtered = len(people_filtered)

print(f"C√¢u tr·∫£ l·ªùi 3: C√≥ {count_filtered} ng∆∞·ªùi th·ªèa m√£n c·∫£ hai ƒëi·ªÅu ki·ªán tr√™n.")
```

**K·∫øt qu·∫£ 3:** C√≥ **4** ng∆∞·ªùi th·ªèa m√£n ƒëi·ªÅu ki·ªán.
